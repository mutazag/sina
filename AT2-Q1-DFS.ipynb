```python
import networkx as nx
import matplotlib.pyplot as plt
from helpers import samples

sample = samples.graph1()
print(sample.nodes)
print(sample.edges)
print(sample.pos)
print(sample.labels)
```

```python
G = nx.Graph()
G.add_nodes_from(sample.nodes)
G.add_edges_from(sample.edges)
```

```python
nx.draw(G)
```

```python
nx.draw_networkx(G, pos=sample.pos,node_size=500,node_color='grey')
```

```python
pos = sample.pos
nx.draw_networkx_nodes(G, pos, node_color='black', node_size=500, alpha=0.8)  #can add nodelist to work with subset of nodes
```

```python

# node_color (color string, or array of floats)  Node color. Can be a single color format string (default=r), or a sequence of colors with the same length as nodelist. If numeric values are specified they will be mapped to colors using the cmap and vmin,vmax parameters. See matplotlib.scatter for more details.


nx.draw_networkx_nodes(G, pos, nodelist=list('ABCD'), node_color='red', node_size=500, alpha=0.8)  #can add nodelist to work with subset of nodes
nx.draw_networkx_nodes(G, pos, nodelist=list('EFGH'), node_color='blue', node_size=500, alpha=0.8)  
nx.draw_networkx_nodes(G, pos, nodelist=list('IJ'), node_color='grey', node_size=500, alpha=0.8)  

#add labels to nodes 
nx.draw_networkx_labels(G, pos, labels=sample.labels, font_size=16)

#add edges 
nx.draw_networkx_edges(G, pos, width=1.5, alpha=0.5, edge_color='g')
```

# implement DFS 

```python
# neighbors of a node 
print(list( G.neighbors('A')))
print(G.degree('A'))

print(list( G.neighbors('F')))
print(G.degree('F'))

```

```python
from queue import Queue

q = Queue()

for n in G.neighbors('A'):
    q.put(n)

print(q.empty())
print(q.queue)
```

```python
while q.empty() is not True:
    print(q.get())
print(q.queue)
```
