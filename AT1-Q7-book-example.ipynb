```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from("ABCDEFGHIJK")
```

```python
edges = [
    ('A','B'),('A','C'),('A','D'),('A','E'), 
    ('B','C'),('B','F'), 
    ('C','F'), 
    ('D','G'),('D','H'), 
    ('E','H'), 
    ('F','I'), 
    ('G','I'),('G','J'), 
    ('H','J'),
    ('I','K'), 
    ('J','K')
]

G.add_edges_from(edges)
```

```python
pos = {
    'E': (1,1), 
    'A': (1,2), 
    'B': (1,3), 
    'H': (2,1), 
    'D': (2,2), 
    'C': (2,3), 
    'F': (2,4), 
    'J': (3,1), 
    'G': (3,2), 
    'I': (3,3), 
    'K': (4,2)
}
```

```python
nx.draw(G, pos=pos, font_color='white', with_labels=True)
```

```python
nx.draw_networkx(G, pos=pos, node_color='grey')
```

```python
A_K_all_shortest_paths = nx.all_shortest_paths(G, source='A',target='K')
[p for p in A_K_all_shortest_paths]
```

```python
A_G_all_shortest_paths = nx.all_shortest_paths(G, source='A',target='G')
[p for p in A_G_all_shortest_paths]
```

```python
A_H_all_shortest_paths = nx.all_shortest_paths(G, source='A',target='H')
[p for p in A_H_all_shortest_paths]
```
