```python
## construct the network
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
G.add_nodes_from("ABCDEFGHI")
G.order()
```

```python
G.add_edges_from([
    ('A','B'),('A','D'),('A','F'),
    ('B','C'),('B','E'),
    ('D','E'),('D','I'),('D','F'), 
    ('F','H'),
    ('C','G'),
    ('E','G'),('E','H'),
    ('H','G'),('H','I'), 
    ('I','G')

])
```

```python
#set positions node: (col -left to right, row -bottom to top)
pos = { 
    'A': (1,2), 
    'B': (2,3), 
    'D': (2,2), 
    'F': (2,1), 
    'C': (3,3), 
    'E': (3,2), 
    'H': (3,1), 
    'G': (4,2.5), 
    'I': (4,1.5)
}
```

```python
nx.draw_networkx(G, pos=pos)
```

```python
from networkx.algorithms.traversal import breadth_first_search

bfs_G_A = breadth_first_search.bfs_tree(G, 'A')
```

```python
nx.draw_networkx(bfs_G_A, )
```
