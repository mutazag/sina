```python
# construct the graph 

import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from("ABCDEFGHIJK")
```

```python
edges = [
    ('A','B'),('A','E'),('A','D'),('A','F'),
    ('B','D'),('B','E'),('B','C'), 
    ('C','E'),('C','I'),('C','G'),
    ('D','F'),('D','H'),('D','E'),('D','G'),
    ('E','I'),('E','H'),('E','G'),('E','F'),
    ('F','H'),('F','J'),('F','G'),
    ('G','H'),('G','K'),('G','I'),
    ('H','J'),('H','K')
    #('I'), 
    #('J'),
    #('K')
]

pos = {
    "A": (4.5,7), 
    "B": (8,7), 
    "C": (10.5,6.5),
    "D": (5,5.5),
    "E": (7.5,4.5), 
    "F": (2,5), 
    "G": (8,2), 
    "H": (4,2.5), 
    "I": (11,4), 
    "J": (1,2), 
    "K": (6.5,1) 
    }
```

```python
G.add_edges_from(edges)
```

```python
# pos = nx.circular_layout(G)
nx.draw_networkx(G, pos, node_size=500, node_color='grey')
```

```python
print(f"Number of nodes: {G.order()}")
print(f"Number of edges: {G.size()}")
```

```python
[c for c in nx.find_cliques(G)]
```
